```python
import pandas as pd

master_df = pd.read_csv(
    "../data/processed/master_deduplicated_leads.csv",
    parse_dates=["created_date", "mql_date", "sql_date", "opportunity_date", "customer_date"]
)

print("Master dataset shape:", master_df.shape)
master_df.head()
```

```python
master_df.info()
```

```python
master_df['lifecycle_stage_final'].value_counts(dropna=False)
```

```python
master_df['is_lead'] = master_df['created_date'].notna()
master_df['is_mql'] = master_df['mql_date'].notna() | (master_df['lifecycle_stage_final'].isin(['MQL', 'SQL', 'Opportunity', 'Customer']))
master_df['is_sql'] = master_df['sql_date'].notna() | (master_df['lifecycle_stage_final'].isin(['SQL', 'Opportunity', 'Customer']))
master_df['is_opportunity'] = master_df['opportunity_date'].notna() | (master_df['lifecycle_stage_final'].isin(['Opportunity', 'Customer']))
master_df['is_customer'] = master_df['customer_date'].notna() | (master_df['lifecycle_stage_final'] == 'Customer')
```

```python
master_df
```

```python
funnel_summary = pd.DataFrame({
    'stage': ['Lead', 'MQL', 'SQL', 'Opportunity', 'Customer'],
    'count': [
        master_df['is_lead'].sum(),
        master_df['is_mql'].sum(),
        master_df['is_sql'].sum(),
        master_df['is_opportunity'].sum(),
        master_df['is_customer'].sum()
    ]
})

funnel_summary
```

### Funnel Progression (Cumulative)
Counts represent leads that reached each stage or beyond.

```python
stage_distribution = master_df['lifecycle_stage_final'].value_counts().reset_index()
stage_distribution.columns = ['stage', 'count']

stage_distribution
```

```python
lead_count = master_df['is_lead'].sum()
mql_count = master_df['is_mql'].sum()
sql_count = master_df['is_sql'].sum()
oppty_count = master_df['is_opportunity'].sum()
customer_count = master_df['is_customer'].sum()

conversion_summary = pd.DataFrame({
    'metric': [
        'Lead to MQL Conversion',
        'MQL to SQL Conversion',
        'SQL to Opportunity Conversion',
        'Opportunity to Customer Conversion',
        'Lead to Customer Conversion'
    ],
    'value': [
        mql_count / lead_count if lead_count else 0,
        sql_count / mql_count if mql_count else 0,
        oppty_count / sql_count if sql_count else 0,
        customer_count / oppty_count if oppty_count else 0,
        customer_count / lead_count if lead_count else 0
    ]
})

conversion_summary['value'] = (conversion_summary['value'] * 100).round(2)

conversion_summary
```

```python
master_df['days_lead_to_mql'] = (master_df['mql_date'] - master_df['created_date']).dt.days
master_df['days_mql_to_sql'] = (master_df['sql_date'] - master_df['mql_date']).dt.days
master_df['days_sql_to_opportunity'] = (master_df['opportunity_date'] - master_df['sql_date']).dt.days
master_df['days_opportunity_to_customer'] = (master_df['customer_date'] - master_df['opportunity_date']).dt.days
```

```python
# Filter valid (non-negative) stage transitions

valid_lead_to_mql = master_df[
    (master_df['days_lead_to_mql'].notna()) &
    (master_df['days_lead_to_mql'] >= 0)
]

valid_mql_to_sql = master_df[
    (master_df['days_mql_to_sql'].notna()) &
    (master_df['days_mql_to_sql'] >= 0)
]

valid_sql_to_opportunity = master_df[
    (master_df['days_sql_to_opportunity'].notna()) &
    (master_df['days_sql_to_opportunity'] >= 0)
]

valid_opportunity_to_customer = master_df[
    (master_df['days_opportunity_to_customer'].notna()) &
    (master_df['days_opportunity_to_customer'] >= 0)
]
```

```python
time_summary = pd.DataFrame({
    'metric': [
        'Lead to MQL Days',
        'MQL to SQL Days',
        'SQL to Opportunity Days',
        'Opportunity to Customer Days'
    ],
    'median_days': [
        valid_lead_to_mql['days_lead_to_mql'].median(),
        valid_mql_to_sql['days_mql_to_sql'].median(),
        valid_sql_to_opportunity['days_sql_to_opportunity'].median(),
        valid_opportunity_to_customer['days_opportunity_to_customer'].median()
    ],
    'avg_days': [
        valid_lead_to_mql['days_lead_to_mql'].mean(),
        valid_mql_to_sql['days_mql_to_sql'].mean(),
        valid_sql_to_opportunity['days_sql_to_opportunity'].mean(),
        valid_opportunity_to_customer['days_opportunity_to_customer'].mean()
    ]
})

time_summary = time_summary.round(2)
time_summary
```

```python
master_df['invalid_mql_before_created'] = master_df['mql_date'] < master_df['created_date']
master_df['invalid_sql_before_mql'] = master_df['sql_date'] < master_df['mql_date']
master_df['invalid_oppty_before_sql'] = master_df['opportunity_date'] < master_df['sql_date']
master_df['invalid_customer_before_oppty'] = master_df['customer_date'] < master_df['opportunity_date']
```

```python
validation_summary = pd.DataFrame({
    'issue': [
        'MQL before Lead Created',
        'SQL before MQL',
        'Opportunity before SQL',
        'Customer before Opportunity'
    ],
    'count': [
        master_df['invalid_mql_before_created'].sum(),
        master_df['invalid_sql_before_mql'].sum(),
        master_df['invalid_oppty_before_sql'].sum(),
        master_df['invalid_customer_before_oppty'].sum()
    ]
})

validation_summary
```

## Data Quality Validation Findings

Validation checks were applied to the master deduplicated funnel to identify impossible stage-date sequences.

### Key findings
- No records had MQL dates before lead creation
- No records had opportunity dates before SQL
- No records had customer dates before opportunity
- 3 records had SQL dates earlier than MQL dates

### Interpretation
These residual sequencing issues reflect manual CRM updates and cross-system backfilling after merger reconciliation. To avoid distortion in time-based KPIs, negative stage durations were excluded from average and median time-to-stage calculations.

```python
master_df['source'] = master_df['source'].str.strip().str.title()
```

```python
source_summary = master_df.groupby('source').agg(
    leads=('dedup_key', 'count'),
    customers=('is_customer', 'sum')
).reset_index()

source_summary['lead_to_customer_pct'] = (
    source_summary['customers'] / source_summary['leads'] * 100
).round(2)

source_summary.sort_values('leads', ascending=False)
```

```python
owner_summary = master_df.groupby('owner').agg(
    leads=('dedup_key', 'count'),
    customers=('is_customer', 'sum')
).reset_index()

owner_summary['lead_to_customer_pct'] = (
    owner_summary['customers'] / owner_summary['leads'] * 100
).round(2)

owner_summary.sort_values('leads', ascending=False)
```
