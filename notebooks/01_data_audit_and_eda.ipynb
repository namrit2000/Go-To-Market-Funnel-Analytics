# Data Audit & Initial Observations – Post-Merger HubSpot Analysis

## Context
This analysis simulates a post-merger scenario where two separate HubSpot instances are used by marketing and sales teams. The goal is to assess data quality, schema consistency, and readiness for unified funnel reporting.

---

## 1.Dataset Overview

- Instance A Shape: 36 rows × 16 columns  
- Instance B Shape: 36 rows × 16 columns  

Both datasets contain similar business information but are structured differently, indicating independent system evolution.

---

## 2.Schema Inconsistencies (Structural Misalignment)

This is a critical issue because both datasets represent the same business entities but use different schemas.

### Key Column Differences:

| Instance A               | Instance B           |
|------------------------|--------------------|
| hs_record_id           | contact_id         |
| email_address          | email              |
| firstname              | first_name         |
| lastname               | last_name          |
| company                | company_name       |
| job_title              | title              |
| lead_source            | original_source    |
| hs_lifecycle_stage     | lifecycle_stage    |
| created_at             | create_date        |
| mql_became_at          | mql_date           |
| sql_became_at          | sal_date           |
| opportunity_created_at | opportunity_date   |
| customer_at            | customer_date      |
| owner_name             | sales_owner        |
| manual_entry_notes     | notes              |
| last_modified_at       | last_touch_date    |

Observation: There is no standardized schema across systems, making direct merging impossible without transformation.

---

## 3.Lifecycle Stage Inconsistencies (Business Logic Issue)

Lifecycle stages are highly inconsistent across both datasets.

### Instance A Values:
- Lead
- MQL / mql
- SQL
- Sales Qualified
- Marketing Qualified Lead
- Opportunity
- Customer

### Instance B Values:
- lead
- MQL
- Marketing Qualified
- SAL
- sql
- Oppty
- Closed Won
- Subscriber

Observations:
- Same stages appear with different naming conventions (e.g., MQL vs Marketing Qualified)
- Case inconsistencies (Lead vs lead vs mql)
- Some stages are ambiguous or misaligned (SAL vs SQL, Closed Won vs Customer)

Impact:
Lifecycle stages cannot be directly trusted for funnel reporting due to inconsistent definitions and naming.

---

## 4.Missing Data (Manual Entry Gaps)

Significant missing values exist in stage progression fields.

### Instance A:
- mql_became_at → 12 missing  
- sql_became_at → 17 missing  
- opportunity_created_at → 25 missing  
- customer_at → 34 missing  

### Instance B:
- mql_date → 15 missing  
- sal_date → 21 missing  
- opportunity_date → 28 missing  
- customer_date → 31 missing  

Observations:
- Later funnel stages have progressively more missing data  
- Customer stage is extremely sparse  

Impact:
Indicates incomplete tracking of lead progression, likely due to manual data entry or lack of enforced processes.

---

## 5. Data Type Issues

All date columns are currently stored as object type instead of datetime.

Impact:
- Cannot perform time-based calculations  
- Cannot validate stage progression  
- Requires conversion before analysis  

---

## 6. Invalid Stage Progression (Data Quality Issue)

Example observed:
- sal_date occurs before mql_date  

Observation:
- Stage transitions are not consistently recorded in chronological order  

Impact:
- Funnel reconstruction based on dates is unreliable without validation  
- Indicates manual updates or delayed data entry  

---

## 7. Duplicate & Identity Issues (Cross-System Overlap)

- No exact duplicates within each dataset  
- Same leads exist across both datasets  
- Variations observed:
  - Email casing differences (uppercase vs lowercase)  
  - Email variations using +coadv  
  - Slight company name differences  

Observation:
This indicates duplicate records across systems.

Impact:
- No unified customer identity  
- Requires deduplication logic after schema alignment  

---

## 8. Funnel Distribution Insight

- Total leads: 36 in each system  
- Very few customers (2–5 records)

Observation:
- Strong drop-off across funnel stages  

Impact:
- Either low conversion rate OR incomplete tracking of final stages  

---

##  Key Takeaway

The primary challenge is not just schema mismatch, but inconsistent lifecycle definitions, unreliable manually entered stage data, and lack of a unified customer identity. These issues prevent accurate funnel reporting and require standardization before any meaningful analysis.

---

##  Next Step

To enable accurate analysis:
- Standardize schema across both datasets  
- Normalize lifecycle stage definitions  
- Convert date fields to proper datetime format  
- Prepare for deduplication and merging
