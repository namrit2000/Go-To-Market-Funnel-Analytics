## 03_merge_and_dedup

```python
import pandas as pd
```

```python
df_a = pd.read_csv(
    "../data/interim/hubspot_a_standardized.csv",
    parse_dates=["created_date", "mql_date", "sql_date", "opportunity_date", "customer_date"]
)

df_b = pd.read_csv(
    "../data/interim/hubspot_b_standardized.csv",
    parse_dates=["created_date", "mql_date", "sql_date", "opportunity_date", "customer_date"]
)

print("A shape:", df_a.shape)
print("B shape:", df_b.shape)
```

```python
df_all = pd.concat([df_a, df_b], ignore_index=True)

print("Combined shape:", df_all.shape)
df_all
```

```python
# Fix lifecycle_stage BEFORE any further logic
df_all['lifecycle_stage'] = df_all['lifecycle_stage'].astype(str).str.strip()

stage_mapping = {
    'Lead': 'Lead',
    'lead': 'Lead',
    'MQL': 'MQL',
    'mql': 'MQL',
    'Marketing Qualified': 'MQL',
    'Marketing Qualified Lead': 'MQL',
    'SQL': 'SQL',
    'sql': 'SQL',
    'Sales Qualified': 'SQL',
    'SAL': 'SQL',
    'Opportunity': 'Opportunity',
    'Oppty': 'Opportunity',
    'Customer': 'Customer',
    'Closed Won': 'Customer',
    'Subscriber': 'Lead',
    'nan': None
}

df_all['lifecycle_stage'] = df_all['lifecycle_stage'].map(stage_mapping)
```

```python
df_all['lifecycle_stage'].isna().sum()
```

```python
duplicate_email_counts = df_all['email'].value_counts()
duplicate_email_counts[duplicate_email_counts > 1]
```

```python
df_all[df_all['email'].isin(duplicate_email_counts[duplicate_email_counts > 1].index)] \
    .sort_values(['email', 'created_date']) \
    [['lead_id', 'email', 'company', 'lifecycle_stage', 'created_date', 'source_system']]
```

```python
df_all['email_base'] = df_all['email'].str.replace(r'\+[^@]+(?=@)', '', regex=True)
```

```python
df_all['company_clean'] = (
    df_all['company']
    .str.lower()
    .str.strip()
    .str.replace(r'[^a-z0-9 ]', '', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
)
```

```python
df_all['full_name_clean'] = (
    df_all['first_name'].str.lower().str.strip() + ' ' +
    df_all['last_name'].str.lower().str.strip()
)
```

```python
email_base_counts = df_all['email_base'].value_counts()
email_base_counts[email_base_counts > 1]
```

```python
df_all[df_all['email_base'].isin(email_base_counts[email_base_counts > 1].index)] \
    .sort_values(['email_base', 'created_date']) \
    [['lead_id', 'email', 'email_base', 'company', 'lifecycle_stage', 'created_date', 'source_system']]
```

```python
name_company_counts = (
    df_all.groupby(['full_name_clean', 'company_clean'])
    .size()
    .reset_index(name='count')
)

name_company_counts[name_company_counts['count'] > 1]
```

```python
possible_dupes = name_company_counts[name_company_counts['count'] > 1][['full_name_clean', 'company_clean']]

df_possible_dupes = df_all.merge(possible_dupes, on=['full_name_clean', 'company_clean'], how='inner')

df_possible_dupes.sort_values(['full_name_clean', 'company_clean', 'created_date'])[
    ['lead_id', 'email', 'full_name_clean', 'company', 'company_clean', 'lifecycle_stage', 'created_date', 'source_system']
]
```

```python
duplicate_email_base_set = set(email_base_counts[email_base_counts > 1].index)

name_company_dup_keys = set(
    tuple(x) for x in name_company_counts[name_company_counts['count'] > 1][['full_name_clean', 'company_clean']].to_records(index=False)
)

df_all['dup_by_email_base'] = df_all['email_base'].isin(duplicate_email_base_set)

df_all['dup_by_name_company'] = df_all.apply(
    lambda row: (row['full_name_clean'], row['company_clean']) in name_company_dup_keys,
    axis=1
)

df_all['is_possible_duplicate'] = df_all['dup_by_email_base'] | df_all['dup_by_name_company']
```

```python
df_all['is_possible_duplicate'].value_counts()
```

```python
df_all['dedup_key'] = df_all['email_base']

name_company_key = df_all['full_name_clean'] + ' | ' + df_all['company_clean']

df_all.loc[
    (~df_all['dup_by_email_base']) & (df_all['dup_by_name_company']),
    'dedup_key'
] = name_company_key
```

```python
stage_rank = {
    'Lead': 1,
    'MQL': 2,
    'SQL': 3,
    'Opportunity': 4,
    'Customer': 5
}

df_all['stage_rank'] = df_all['lifecycle_stage'].map(stage_rank)
```

```python
def first_non_null(series):
    non_null = series.dropna()
    return non_null.iloc[0] if len(non_null) > 0 else pd.NA

def combine_source_systems(series):
    return " | ".join(sorted(series.dropna().astype(str).unique()))

def get_highest_stage(stage_series):
    valid = stage_series.dropna()
    if valid.empty:
        return pd.NA
    return max(valid, key=lambda x: stage_rank.get(x, 0))
```

```python
master_df = df_all.groupby('dedup_key').agg({
    'lead_id': first_non_null,
    'email_base': first_non_null,
    'email': first_non_null,
    'first_name': first_non_null,
    'last_name': first_non_null,
    'company': first_non_null,
    'job_title': first_non_null,
    'source': first_non_null,
    'created_date': 'min',
    'mql_date': 'min',
    'sql_date': 'min',
    'opportunity_date': 'min',
    'customer_date': 'min',
    'owner': first_non_null,
    'notes': first_non_null,
    'last_updated': first_non_null,
    'source_system': combine_source_systems
}).reset_index()
```

```python
stage_df = (
    df_all.groupby('dedup_key')['lifecycle_stage']
    .apply(get_highest_stage)
    .reset_index(name='lifecycle_stage_final')
)

master_df = master_df.merge(stage_df, on='dedup_key', how='left')
```

```python
counts = df_all.groupby('dedup_key').size().reset_index(name='records_merged')

master_df = master_df.merge(counts, on='dedup_key', how='left')
```

```python
master_df = master_df[
    [
        'dedup_key',
        'lead_id',
        'email_base',
        'email',
        'first_name',
        'last_name',
        'company',
        'job_title',
        'source',
        'lifecycle_stage_final',
        'created_date',
        'mql_date',
        'sql_date',
        'opportunity_date',
        'customer_date',
        'owner',
        'notes',
        'last_updated',
        'source_system',
        'records_merged'
    ]
]
```

## Deduplication Strategy

To create a unified lead record across merged HubSpot instances:

1. Identify duplicates using:
   - Normalized email (email_base)
   - Name + company combination
   
  
2. For each duplicate group:
   - Keep earliest created_date (first touch)
   - Keep latest lifecycle_stage (most advanced stage)
   - Combine stage dates by taking non-null values
   - Prefer non-null values across fields (coalesce logic)

3. Treat each group as one unique lead

```python
# Start with normalized email as default group key
df_all['dedup_key'] = df_all['email_base']

# For records not caught well by email pattern differences,
# use name + company as fallback grouping signal
name_company_key = df_all['full_name_clean'] + ' | ' + df_all['company_clean']

# If a record is duplicate by name/company but not by email_base,
# use the name-company key
df_all.loc[
    (~df_all['dup_by_email_base']) & (df_all['dup_by_name_company']),
    'dedup_key'
] = name_company_key
```

```python
stage_rank = {
    'Lead': 1,
    'MQL': 2,
    'SQL': 3,
    'Opportunity': 4,
    'Customer': 5
}

df_all['stage_rank'] = df_all['lifecycle_stage'].map(stage_rank)
```

```python
def get_max_stage(series):
    valid_stages = series.dropna()
    if valid_stages.empty:
        return pd.NA
    
    return valid_stages.map(stage_rank).idxmax()
```

```python
def get_highest_stage(stage_series):
    valid_stages = stage_series.dropna()
    if valid_stages.empty:
        return pd.NA
    
    highest_stage = max(valid_stages, key=lambda x: stage_rank.get(x, 0))
    return highest_stage
```

```python
def first_non_null(series):
    non_null = series.dropna()
    return non_null.iloc[0] if len(non_null) > 0 else pd.NA

def combine_source_systems(series):
    return " | ".join(sorted(series.dropna().astype(str).unique()))

master_df = df_all.groupby('dedup_key').agg({
    'lead_id': first_non_null,
    'email_base': first_non_null,
    'email': first_non_null,
    'first_name': first_non_null,
    'last_name': first_non_null,
    'company': first_non_null,
    'job_title': first_non_null,
    'source': first_non_null,
    'created_date': 'min',
    'mql_date': 'min',
    'sql_date': 'min',
    'opportunity_date': 'min',
    'customer_date': 'min',
    'owner': first_non_null,
    'notes': first_non_null,
    'last_updated': first_non_null,
    'source_system': combine_source_systems
}).reset_index()
```

```python
highest_stage_df = (
    df_all.groupby('dedup_key')['lifecycle_stage']
    .apply(get_highest_stage)
    .reset_index(name='lifecycle_stage_final')
)

master_df = master_df.merge(highest_stage_df, on='dedup_key', how='left')
```

```python
group_sizes = (
    df_all.groupby('dedup_key')
    .size()
    .reset_index(name='records_merged')
)

master_df = master_df.merge(group_sizes, on='dedup_key', how='left')
```

```python
master_df = master_df[
    [
        'dedup_key',
        'lead_id',
        'email_base',
        'email',
        'first_name',
        'last_name',
        'company',
        'job_title',
        'source',
        'lifecycle_stage_final',
        'created_date',
        'mql_date',
        'sql_date',
        'opportunity_date',
        'customer_date',
        'owner',
        'notes',
        'last_updated',
        'source_system',
        'records_merged'
    ]
]
```
